```python
import tensorflow as tf
from keras.models import  Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
import pickle
import time
from tensorflow.keras.callbacks import TensorBoard
```

```python
NAME = "Cats-vs-dogs-CNN-{}".format(int(time.time()))
```

```python
pickle_in = open("X.pickle","rb")
X = pickle.load(pickle_in)

pickle_in = open("y.pickle","rb")
y = pickle.load(pickle_in)

```

```python
X = X/255.0
print(X)
```

```python
X.shape[1:]
```

```python
model = Sequential()

model.add(Conv2D(256, (3, 3), input_shape=X.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(256, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())  # this converts our 3D feature maps to 1D feature vectors


model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
tensorboard = TensorBoard(log_dir="logs/{}".format(NAME))
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(X, y, batch_size=32, epochs=5, validation_split=0.3, callbacks = [tensorboard])
# to access tensorboard ==> at cmd type => tensorboard --logdir==training:logs/ --host=127.0.0.1
```
